```python
import pandas as pd
```

```python
df=pd.read_csv('Customer Purchasing Behaviors.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Le=LabelEncoder()
```

```python
df['region']=Le.fit_transform(df['region'])
```

```python
df
```

```python
X=df.drop(columns=['user_id','loyalty_score'])
```

```python
y=df['loyalty_score']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg=DecisionTreeRegressor(max_depth=7)
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score(X_train,y_train)
```

```python
reg.score(X_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rareg=RandomForestRegressor()
```

```python
rareg.fit(X_train,y_train)
```

```python
rareg.score(X_train,y_train)
```

```python
rareg.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error,mean_absolute_percentage_error,max_error,explained_variance_score,accuracy_score
```

```python
y_pred=rareg.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
Rmse=mean_squared_error(y_test,y_pred,squared=False)
```

```python
Rmse
```

```python
mae=mean_absolute_error(y_test,y_pred)
```

```python
mae
```

```python
MAPE=mean_absolute_percentage_error(y_test,y_pred)
```

```python
MAPE
```

```python
print(f"mean absolute percentage error is {MAPE*100:.2f}%")
```

```python
evs=explained_variance_score(y_test,y_pred)
```

```python
evs
```
